```python
import pandas as pd
import scipy
from scipy import stats
import statsmodels.api as sm
import numpy as np
from statsmodels.formula.api import ols
from statsmodels.stats.proportion import proportions_ztest
```

```python
cutlet=pd.read_csv("Cutlets.csv")
```

```python
cutlet
```

```python
count=pd.crosstab(cutlet['Unit A'],cutlet['Unit B'])
```

```python
count
```

```python
cutlet.columns="UnitA","UnitB"
```

```python
print(stats.shapiro(cutlet.UnitA))
```

```python
print(stats.shapiro(cutlet.UnitB))
```

```python
scipy.stats.levene(cutlet.UnitA,cutlet.UnitB)
```

```python
sm.stats.ttest_ind(cutlet.UnitA,cutlet.UnitB)
```

```python
sm.stats.ttest_ind(cutlet.UnitA,cutlet.UnitB,alternative='smaller')
```
